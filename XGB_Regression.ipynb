```python
#importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
df = pd.read_csv(r"housing.csv")
```

```python
oceanpro = df["ocean_proximity"]
```

```python
# one hot encode the color column
one_hot_encoded_data = pd.get_dummies(df, columns=['ocean_proximity'])
print(one_hot_encoded_data)
```

```python
df
```

```python
one_hot_encoded_data
```

```python
one_hot_encoded_data = one_hot_encoded_data.replace(False, 0)
one_hot_encoded_data = one_hot_encoded_data.replace(True, 1)
one_hot_encoded_data
```

```python
df = one_hot_encoded_data
```

```python
y = df["median_house_value"]
```

```python
df.drop(columns=['median_house_value'])
```

```python
x = df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
df.info()
```

```python
df.rename(columns={'ocean_proximity_<1H OCEAN': 'ocean_proximity_LESS_THAN_AN_HOUR'}, inplace=True)
```

```python
model = xgb.XGBRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, predictions)
```

```python
print("Mean Squared Error: ", mse)
```

```python
r_square = r2_score(y_test, predictions)
```

```python
print("R Squared score: ", r_square)
```
